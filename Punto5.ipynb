```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def metodo_euler(alpha, u0, dt, pasos):
    u_values = [u0]

    for _ in range(pasos):
        u_n = u_values[-1]
        u_next = u_n * (1 + alpha * dt)
        u_values.append(u_next)

    return u_values
```

```python
# parametros
alpha = 0.1
u0 = 1.0
dt = 0.1
pasos = 10
resultado_euler = metodo_euler(alpha, u0, dt, pasos)
resultado_exacto = [(1 + alpha * dt)**k * u0 for k in range(pasos + 1)]
```

Se demuestra que con el metodo de euler se obtiene la misma solución

```python

plt.plot(range(pasos + 1), resultado_euler, label='Método de Euler')
plt.plot(range(pasos + 1), resultado_exacto, label='Solución Exacta', linestyle='--')

plt.xlabel('Iteraciones')
plt.ylabel('u(t)')
plt.title('Método de Euler vs Solución Exacta')
plt.legend()

plt.show()
```

```python
#parametros de los puntos
alpha = -1
u0 = 1.0
delta_ts = [1.1, 1.5, 1.9]
```

```python

fig, axs = plt.subplots(len(delta_ts), figsize=(8, 6), sharex=True)
for i, dt in enumerate(delta_ts):
    pasos = int(5 / dt)
    resultado_euler = metodo_euler(alpha, u0, dt, pasos)

    # grfico para cada dt
    axs[i].plot(np.arange(0, (pasos + 1) * dt, dt), resultado_euler, label=f'dt = {dt}')
    axs[i].set_ylabel('u(t)')
    axs[i].legend()
plt.xlabel('Tiempo (s)')
plt.suptitle('Soluciones Oscilatorias con α = -1 y Diferentes Δt')
plt.show()
```
