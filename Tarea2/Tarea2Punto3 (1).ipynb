```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
h = 0.005
#t=np.linspace(0,1+h,50)
#t = np.arange(0,1+h,h)
#t
x0 = np.sqrt(2)
b = 4 
y0 = 0 
dy0 = -2 / np.sqrt(2)  # Valor inicial de y'(x)
h = 0.01
```

```python
def f(x, y, u):
    return (x**4 * y**2 - 2 * x**2 * y - 1) / x**3
```

```python
def GetEuler3(f, x0, b, y0, u0, h):
    n = int((b - x0) / h)
    x = np.linspace(x0, b, n + 1)
    y = np.zeros(n + 1)
    dy = np.zeros(n + 1)
    y[0] = y0
    dy[0] = dy0

    for i in range(n):
        k1 = h * dy[i]
        y1 = h * f(x[i], y[i], dy[i])
        k2 = h * (dy[i] + 0.5 * y1)
        y2 = h * f(x[i] + 0.5 * h, y[i] + 0.5 * k1, dy[i] + 0.5 * y1)
        k3 = h * (dy[i] - y1 + 2 * y2)
        y3 = h * f(x[i] + h, y[i] - k1 + 2 * k2, dy[i] - y1 + 2 * y2)
        
        y[i + 1] = y[i] + (k1 + 4 * k2 + k3) / 6
        dy[i + 1] = dy[i] + (y1 + 4 * y2 + y3) / 6

    return x, y
#print("Valores de x:")
#print(x)
#print("Valores de y:")
#print(y)
```

```python
#t= np.linspace(0, 6, 40)
x, y = GetEuler3(f,x0,b,y0,dy0,h)
```
