Operadores diferenciales:

\begin{equation}
f^\prime\left(x\right)=\frac{-f\left(x+2h\right)+4f\left(x+h\right)-3f\left(x\right)}{2h}
\end{equation}

\begin{equation}
f\prime\prime(x)\ =\ \frac{f(x+h)\ -\ 2f(x)\ +\ f(x-h)}{h^2}\ 
\end{equation}

La primera y segunda derivada de $x^2$ son 2x y 2, respectivamente. De esta manera, usandolo como función de prueba se tiene lo siguiente:

\begin{equation}
f^\prime\left(x\right)=\frac{-\left(x+2h\right)^2+4\left(x+h\right)^2-3\left(x\right)^2}{2h}
=\frac{-\left(x^2+4hx+4h^2\right)^+4\left(x^2+2xh+h^2\right)^-3\left(x\right)^2}{2h}
\end{equation}

Se suman y restan los términos, dando como resultado:

\begin{equation}
=\frac{-4hx-4h^2+4x^2+8xh+{4h}^2-4\left(x\right)^2}{2h}=\frac{-4hx+8xh}{2h}=\frac{4xh}{2h}=2x
\end{equation}

De esta manera, se probó que el primer operador diferencial era consistente, ya que da el mismo resultado.

\begin{equation}
f\prime\prime(x)\ =\ \frac{(x+h)^2\ -\ 2x^2\ +\ (x-h)^2}{h^2}
\end{equation}

\begin{equation}
f^{\prime\prime}\left(x\right)=\frac{x^2+2xh+h^2-2x^2+x^2-2xh+h^2}{h^2}=\frac{2h^2}{h^2}=2
\end{equation}

De esta manera, se probó que el segundo operador diferencial era consistente, ya que da el mismo resultado que la segunda deriva.


Para probar con f(x) = sin(x) lo mejr es utilizar codigo



```python
import numpy as np
import matplotlib.pyplot as plt

def fp(f, x, h):
    return (-f(x + 2*h) + 4*f(x + h) - 3*f(x)) / (2*h)

def fdoublep(f, x, h):
    return (f(x + h) - 2*f(x) + f(x - h)) / (h**2)

def f(x):
    return np.sin(x)
```

```python
x = np.linspace(0, 6, 100)
#paso
h=0.01
fp_=fp(f,x,h)
fdoublep_=fdoublep(f,x,h)
ex_der=np.cos(x)
ex_dder=-np.sin(x)

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(x, fp_ , label='Operador')
plt.plot(x, ex_der, '--', label='Derivada exacta')
plt.title('Primera derivada')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x, fdoublep_, label='Aproximación numérica')
plt.plot(x, ex_dder, '--', label='Derivada exacta')
plt.title('Segunda derivada')
plt.legend()
```
